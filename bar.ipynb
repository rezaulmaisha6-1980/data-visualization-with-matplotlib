#Bar chart 

```python
import matplotlib.pyplot as plt
import numpy as np 
 
class_name = ["One", "Two", "Three", "Four", "Five", "Six", "Seven", "Eight"]

class_student = [45, 34, 50, 42, 35, 55, 41, 35]

x = np.arange(len(class_name))

# width = 0.50

plt.bar( class_name, class_student,  color='blue', label='Student information of class') # add width 
plt.title( "Student information bar chart")
plt.xlabel('Class')
plt.ylabel('Student')

plt.legend()
plt.show()
```
