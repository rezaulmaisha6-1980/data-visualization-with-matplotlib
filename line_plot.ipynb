#Line plot 

```python
import matplotlib.pyplot as plt

month = ["January", "February", "March", "May", "June", "July"]

sales2023 = [ 20, 15, 18, 22, 17, 25 ]
sales2024 = [ 24, 20, 15, 19, 11, 22]
sales2025 = [18, 13,22, 12, 16, 14]


plt.plot(month, sales2023, color='green', marker="o", linestyle='--' , label='sales 2023',linewidth=1, markersize=5 )
plt.plot(month, sales2024, color='red', marker="s", linestyle='--' , label='sales 2024', linewidth=1, markersize=5 )
plt.plot(month, sales2025, color='blue', marker="o", linestyle='--' , label='sales 2025',linewidth=1, markersize=5  )
plt.title(" Monthly sales of Oil")
plt.xlabel('Months')
plt.ylabel("Sales/lakh(Taka)")
plt.legend()
plt.show()




```
