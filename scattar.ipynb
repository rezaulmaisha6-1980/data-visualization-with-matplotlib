```python

import matplotlib.pyplot as plt


student_number = [5, 8, 7, 9, 10, 13]
math_number = [ 99, 80, 66, 50, 45, 33]

plt.scatter (student_number, math_number, color='red')
plt.title(" Math number graph ")

plt.xlabel(" Student ")

plt.ylabel(" Number of math")

plt.show()
```
