#Histogram chart 

```python
import matplotlib.pyplot as plt

ages = [22, 23, 30, 28, 30, 22, 40, 28, 42, 30]

plt.hist(ages, bins=5, color="blue", edgecolor="black", cumulative=True)

plt.title("Age Distribution")

plt.xlabel("Age")

plt.ylabel("Frequency")

plt.show()
```

```python


x = [22, 25, 30, 22, 28]

y = [30, 35, 28, 40, 42]

plt.hist([x, y], bins=5, color=["red", "blue"], label=['Product-A', 'Product-B'])

plt.title("Product Distribution Comparison")

plt.xlabel("Product")

plt.ylabel("Frequency")

plt.legend( )

plt.show()
```
